```python
# Day 1 Challenge
#
# Using math operators in Python, find out which flight will give Dot 
# the best value per hour of travel time.
#
# value = price_paid / travel_time
#
# Revision History
#     Tonnicca Gelacio, 2022-04-12 : Created
#     Tonnicca Gelacio, 2022-04-12 : Code Updated
#
```

```python
def main():
    # Declarations - First Leg Flight Prices
    van_tor_price = 250
    van_ott_price = 280
    van_mon_price = 240
    van_edm_price = 150
    van_cal_price = 180

    # Declarations - First Leg Travel Time
    van_tor_travel_time = 3.5
    van_ott_travel_time = 4
    van_mon_travel_time = 4
    van_edm_travel_time = 1.5
    van_cal_travel_time = 1

    # Declarations - Second Leg Flight Prices
    ott_ber_price = 1350
    mon_lon_price = 1300
    edm_lon_price = 1290
    cal_lon_price = 1400
    tor_mun_price = 990
    # edm_lon_price = 1200

    # Declarations - Layover
    ott_layover = 3.5
    mon_layover = 2
    edm_layover = 5
    cal_layover = 2.5
    tor_layover = 1.5

    # Declarations - Second Leg Travel Time
    ott_ber_travel_time = 9
    mon_lon_travel_time = 8
    edm_lon_travel_time = 10
    cal_lon_travel_time = 10
    tor_mun_travel_time = 9.5

    # Vancouver -> Toronto -> Munich
    vtm_value = round((van_tor_price + tor_mun_price) / 
        (van_tor_travel_time + tor_layover + tor_mun_travel_time), 2)
    print(f'Vancouver -> Toronto -> Munich: {vtm_value} CAD')

    # Vancouver -> Ottawa -> Berlin
    vob_value = round((van_ott_price + ott_ber_price) / 
        (van_ott_travel_time + ott_layover + ott_ber_travel_time), 2)
    print(f'Vancouver -> Ottawa -> Berlin: {vob_value} CAD')

    # Vancouver -> Montreal -> London
    vml_value = round((van_mon_price + mon_lon_price) / 
        (van_mon_travel_time + mon_layover + mon_lon_travel_time), 2)
    print(f'Vancouver -> Montreal -> London: {vml_value} CAD')

    # Vancouver -> Edmonton -> London
    vel_value = round((van_edm_price + edm_lon_price) / 
        (van_edm_travel_time + edm_layover + edm_lon_travel_time), 2)
    print(f'Vancouver -> Edmonton -> London: {vel_value} CAD')

    # Vancouver -> Calgary -> London
    vcl_value = round((van_cal_price + cal_lon_price) / 
        (van_cal_travel_time + cal_layover + cal_lon_travel_time), 2)
    print(f'Vancouver -> Calgary -> London: {vcl_value} CAD')

    # create list to store flight details and value
    flight_values = [['Vancouver -> Toronto -> Munich', vtm_value], 
        ['Vancouver -> Ottawa -> Berlin', vob_value],
        ['Vancouver -> Montreal -> London', vml_value],
        ['Vancouver -> Edmonton -> London', vel_value],
        ['Vancouver -> Calgary -> London', vcl_value]]

    # get best value using min()
    best_label, best_value= min(flight_values, key=lambda i: i[1])

    # print result
    print(f'Best value per hour of travel time: {best_label} (Value: {best_value} CAD)')

if __name__ == "__main__":
    main()
```

```python
### OUTPUT ###
# Vancouver -> Toronto -> Munich: 85.52 CAD
# Vancouver -> Ottawa -> Berlin: 98.79 CAD
# Vancouver -> Montreal -> London: 110.0 CAD
# Vancouver -> Edmonton -> London: 87.27 CAD
# Vancouver -> Calgary -> London: 117.04 CAD
# Best value per hour of travel time: Vancouver -> Toronto -> Munich (Value: 85.52 CAD)
```
